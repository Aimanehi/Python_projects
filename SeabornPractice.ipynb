### Seaborn
Seaborn is a python data visualization library based on mathplotlib.
It provides a high leve interface for drawing attractive and informative statistical graphics 

- Distribution Plot
- Categorical Plot 
- Regression Plot

#### Distribution Plot
- Distplot
- Jointplot
- PairPlot


```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df.head() #the first five rows in the data
```

```python
# Distplot
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'], kde=False, bins=20)
```

```python
# joint plots - Scatter, Reg, Resid, kde, hex
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='resid')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
# pair plot---- plot all relatioship available in the data

sns.pairplot(df)
```

#### Categorical Plot
- factorplot
- boxplot
- barplot
- countplot
- vioinplot
- swarmplot
- stripplot

```python
df.head()
```

```python
#factor plot-- it is the generic visualization

sns.factorplot(x='sex', y='total_bill', data=df, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=df, kind='bar')
```

```python
# countplot - based on a count
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='day', data=df, hue='sex')
```

```python
# barplot
sns.barplot(x='sex',y='total_bill', data=df)
```

```python
# boxplot
sns.boxplot(x='sex',y='tip', data=df)
```

```python
sns.boxplot(x='sex',y='total_bill', data=df, palette='magma')
```

```python
# regression plot --lmplot
df.head()
```

```python
sns.lmplot(x='size',y='tip', data=df)
```

```python
sns.lmplot(x='total_bill',y='tip', data=df, hue='sex',palette='magma_r')
```
