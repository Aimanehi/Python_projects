### Data Visualization in Python

#### Matplotlib

Matplotlib is a comprehensive library for creating static, animated and interactive visualization in Python

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
header = ['name', 'age', 'quantity', 'revenue']
df = [['Ben', 23, 250, 2500],['Kerry', 21, 30, 4590],['Micheal', 24, 100, 1500],['Nene', 24, 90, 1700],['James', 27, 800, 3000]]
```

```python
df = pd.DataFrame(data=df, columns=header)
df
```

```python
# pit.plot(x,y) - (name, age)

plt.title("Customer Age")
plt.plot(df['name'],df['age'])
plt.show()
```

```python
plt.title("Total Quantity by Customer")
plt.bar(df['name'],df['quantity'])
plt.show()
```

```python
df.sort_values('quantity', inplace=True, ascending=True)
```

```python
plt.title("Total Quantity by Customer")
plt.bar(df['name'],df['quantity'])
plt.show()
```

```python
# Bar Chart
plt.title('Total Revenue by Customer')
df.sort_values('revenue', inplace=True, ascending=True)
plt.barh(df['name'], df['revenue'])
plt.show()
```

```python
# relationship between age and revenue
plt.title('Realtionship between customer age and revenue')
plt.scatter(x='age', y='revenue', data=df)
plt.show()
```

```python
# relationship between name and revenue
plt.title('Realtionship between customer age and revenue')
plt.pie('revenue', labels='name', data=df)
plt.show()
```

### Subplots()

```python
fig, axs = plt.subplots()
```

```python
fig, axs = plt.subplots(3,2, figsize=(10,10))

axs[0,0].set_title('Customer Age Analysis')
axs[0,0].bar(df['name'], df['age'])

axs[0,1].set_title('Total Quantity by Customer')
axs[0,1].barh(df['name'], df['quantity'])

axs[1,0].set_title('Total Revenue by Customer')
axs[1,0].plot(df['name'], df['revenue'])

axs[1,1].set_title('Realtionship between Age and Revenue')
axs[1,1].scatter(df['age'], df['revenue'])

axs[2,0].set_title('Realtionship between customer age and revenue')
axs[2,0].pie('revenue', labels='name', data=df)

axs[2,1].set_title('Total Quantity by Customer')
#axs[2,1].hist(df['name'], df['quantity'])
axs[2,1].hist('age', data=df)

plt.show()
```
